```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp as ivp
```

## Ejemplo 5.1.1

```python
def a_os(t,z,omega):
  dz = [z[1],-z[0]*omega**2]
  return dz

fg_time, ax_time = plt.subplots()
fg_ss, ax_ss = plt.subplots()

tend = 10*np.pi
omega = 1
for z_01 in np.linspace(0.5,1.5,3):
  z0 = [z_01,0]
  sol = ivp(a_os, [0,tend], z0, args=(omega,), dense_output=True)

  t = np.linspace(0,tend,10000)
  z = sol.sol(t)

  ax_time.plot(t,z[0,:])
  ax_time.plot(t,z[1,:])

  ax_ss.set_title('Diagrama de fase - Espacio de estados')
  ax_ss.plot(z[0,:],z[1,:])
```

## Ejemplo:

* _22/11/2021_

$$\dot{x_1} = x_1 + e^{-x_2}$$
$$\dot{x_2} = -x_2$$

```python
def a_os(t,z,omega):
  dz = [z[0]+np.exp(-z[1]),-z[1]]
  return dz

fg_time, ax_time = plt.subplots()
fg_ss, ax_ss = plt.subplots()

tend = 1
omega = 1

for z_01 in np.linspace(-2,2,50):
  z0 = [z_01,0]
  sol = ivp(a_os, [0,tend], z0, args=(omega,), dense_output=True)

  t = np.linspace(-tend,tend,10000)
  z = sol.sol(t)

  ax_time.axvline(x=-1, color='black')
  ax_time.plot(t,z[0,:])
  ax_time.plot(t,z[1,:])
  ax_time.grid('on')
  ax_time.text(x=-1.6, y=3,s="(-1,0)")

  ax_ss.set_title('Diagrama de fase - Espacio de estados')
  ax_ss.plot(z[0,:],z[1,:])
  plt.grid('on')
```

## Ejemplo:

$$\dot{x} = -x + ay +x^2y$$
$$\dot{y} = b - ay -x^2y$$

### Stable Limit Cycle

```python
def a_os(t,z,a, b):
  dz = [-z[0] + a*z[1] + z[0]**2 * z[1], b - a*z[1] - z[0]**2 * z[1]]
  return dz

fg_time, ax_time = plt.subplots()
fg_ss, ax_ss = plt.subplots()

tend = 50
a = 0.1
b = 0.5
for z_01 in np.linspace(-1,2.5,7):
  for z_02 in np.linspace(-1,2.5,7):
    z0 = [z_01,z_02]
    sol = ivp(a_os, [0,tend], z0, args=(a,b,), dense_output=True)

    t = np.linspace(0,tend,10000)
    z = sol.sol(t)

    ax_time.plot(t,z[0,:])
    ax_time.plot(t,z[1,:])
    ax_time.grid('on')

    ax_ss.set_title('Diagrama de fase - Espacio de estados')
    ax_ss.plot(z[0,:],z[1,:])
    ax_ss.grid('on')
```

```python
def a_os(t,z,a, b):
  dz = [-z[0] + a*z[1] + z[0]**2 * z[1], b - a*z[1] - z[0]**2 * z[1]]
  return dz

fg_time, ax_time = plt.subplots()
fg_ss, ax_ss = plt.subplots()

tend = 50
omega = 1
a = 0.06
b = 1.2
for z_01 in np.linspace(-1,2.5,7):
  for z_02 in np.linspace(-1,2.5,7):
    z0 = [z_01,z_02]
    sol = ivp(a_os, [0,tend], z0, args=(a,b,), dense_output=True)

    t = np.linspace(0,tend,10000)
    z = sol.sol(t)

    ax_time.plot(t,z[0,:])
    ax_time.plot(t,z[1,:])
    ax_time.grid('on')

    ax_ss.set_title('Diagrama de fase - Espacio de estados')
    ax_ss.plot(z[0,:],z[1,:])
    ax_ss.grid('on')
```

## Ejemplo: 

### Van der Pol Oscilator

```python
def a_os(t,z,mu):
  dz = [z[1], mu*(1-z[0]**2)*z[1]-z[0]]
  return dz

fg_time, ax_time = plt.subplots()
fg_ss, ax_ss = plt.subplots()

tend = 20
mus = [0,1,3]
styles=["-","--",":"]
for mu,style in zip(mus,styles):
  sol=ivp(a_os, [0,tend], [0,1], args=(mu,), dense_output=True)
  t = np.linspace(0,tend,10000)
  z = sol.sol(t)

  ax_time.plot(t,z[0,:], style)
  #ax_time.plot(t,z[1,:], style)
  ax_time.grid('on')

  ax_ss.set_title('Diagrama de fase - Espacio de estados')
  ax_ss.plot(z[0,:],z[1,:], style)
  ax_ss.grid('on')
```
