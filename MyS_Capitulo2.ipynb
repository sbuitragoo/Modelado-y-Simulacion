## Libraries and Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler_method(f, h, x0, ti, tf):
  N = int((tf-ti)/h)+1
  x = np.zeros(N)
  x[0] = x0
  for i in range(len(x)-1):
    x[i+1] = x[i] + h*f(x[i])

  t = np.linspace(ti,tf,N)
  
  return t,x
```

Analyze the following equations graphically. ### 2.2.1

$$\dot{x} = 4x^2 -16 $$

```python
def function(t):
  return (4*t**2) - 16

h = 0.001
x0 = 1
ti = 0
tf = 0.1
x0_i = -3
x0_f = 3
n = 8

for i in np.linspace(x0_i,x0_f,n):
    x0 = i
    t,x = euler_method(function,h,x0,ti,tf)
    plt.plot(t,x,label=f'$x_0 = {i}$')
    plt.grid('on')
    plt.xlabel('$t$')
    plt.ylabel('$x(t)$')
    plt.ylim(-4,10)
```
