```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp as solver
from ipywidgets import interact
```

#5.3 Love Affairs

## 5.3.1
Suggest names for the four romantic styles, determined by the signs of a and be in $\dot R = aR + bJ$

```python
def f(t,y,a,b):
  R,J = y
  R_dot = a*R + b*J
  J_dot = b*R + a*J

  return R_dot, J_dot

t_span = [-0.1,0.2]

@interact(a=(-10,10),b=(-10,10))
def intreactive(a = 1,b=1):
  plt.figure(figsize=(16,6))
  parameters = [a,b]
  for R in range(-5,5):
    for J in range(-5,5):
      sol = solver(f,t_span=t_span,y0=[R,J],args=parameters)
      plt.plot(sol.y[0],sol.y[1])
      plt.xlabel("$R$",fontsize=20)
      plt.ylabel("$J$",fontsize=20)
      plt.title(f"$a = {a}$, $b = {b}$\n$a ^ 2 = {a**2}$, $b ^ 2 = {b**2}$",fontsize=20)
      plt.grid('on')
  plt.axis([-20,20,-20,20])
  plt.show()
```

## 5.3.2
Consider the affair described by:

$$\dot{R} = J$$

$$\dot{J} = -R + J$$

```python
def f(t,y):
  R,J = y
  R_dot = J
  J_dot = -R + J
  return R_dot, J_dot


t_span = [0,15]
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
for R0 in range(-2,2):
  for J0 in range(-2,2): 
      y0 = [R0,J0]
      sol = solver(f, t_span = t_span, y0 = y0)

      plt.plot(sol.y[0],sol.y[1])
      plt.title("Vector Field")
      plt.xlabel("$R$")
      plt.ylabel("$J$")
      plt.grid('on')
sol = solver(f, t_span = t_span, y0 = [1,0])
plt.subplot(1,2,2)

plt.plot(sol.t,sol.y[0],label="$R(t)$")
plt.plot(sol.t,sol.y[1],label="$J(t)$")
plt.title("$R(t)$ and $J(t)$")
plt.xlabel("$t$")
plt.ylabel("Love/Hate")
plt.ylim(-300,300)
plt.grid('on')

plt.legend()
plt.tight_layout()
plt.show()
```

## 5.3.3
(Out of touch with their feelings)

$$\dot{R} = aJ$$

$$\dot{J} = bR$$

```python
def f(t,y,a,b):
  R,J = y
  R_dot = a*J
  J_dot = b*R
  return R_dot, J_dot

test_values = [[-3,-6],[4,-3],[-5,4],[3,6]]
ti,tf = 0,3
t=np.arange(ti,tf,0.001)

plt.figure(figsize=(30,9))
i = 1
for parameters in test_values:
  plt.subplot(1,4,i)
  i+=1
  for R0 in range(-5,5):
    for J0 in range(-5,5):
      sol3 = solver(f,t_span = [ti,tf],y0=[R0,J0],args=parameters,t_eval=t)
      plt.plot(sol3.y[0],sol3.y[1])
      plt.grid('on')
      title=f"$a = {parameters[0]}$, $b = {parameters[1]}$"
      a,b=parameters
      if (a>0):
        title=title+f"\n$a > 0$"
      else:
        title = title+f"\n$a < 0$"
      if (b<0):
        title = title+f"\n$b<0$"
      else:
        title = title+f"\n$b>0$"
      plt.title(title)
      plt.axis([-10,10,-10,10])
```

## 5.3.4
(Fire and water) Do opposites attract? Analyze:

$$\dot{R} = aR + bJ$$

$$\dot{J} = -bR - aJ$$

```python
def f(t,y,a,b):
  R,J = y
  R_dot = a*R + b*J
  J_dot = -b*R - a*J
  return R_dot, J_dot

test_values = [[-3,-6],[4,-3],[-5,4],[3,6]]
ti,tf = 0,3
t=np.arange(ti,tf,0.001)

plt.figure(figsize=(30,9))
i = 1
for parameters in test_values:
  plt.subplot(1,4,i)
  i+=1
  for R0 in range(-5,5):
    for J0 in range(-5,5):
      sol3 = solver(f,t_span = [ti,tf],y0=[R0,J0],args=parameters,t_eval=t)
      plt.plot(sol3.y[0],sol3.y[1])
      plt.grid('on')
      title=f"$a = {parameters[0]}$, $b = {parameters[1]}$"
      a,b=parameters
      if (a>0):
        title=title+f"\n$a > 0$"
      else:
        title = title+f"\n$a < 0$"
      if (b<0):
        title = title+f"\n$b<0$"
      else:
        title = title+f"\n$b>0$"
      plt.title(title)
      plt.axis([-10,10,-10,10])
```
